## Operaciones lógicas con perceptrones

```python
import numpy as np


#Definir funcion de activación
def activation(value):
    if value<0:
        return 0
    else:
        return 1
    
#Función de multiplicación de matirces
def neural_network(inputs, weights):
    return np.dot(inputs,weights)

bias=1



#Perceptrón para la compuerta AND
def and_perceptron(x1, x2,prnt=True):
    W_AND=np.array([[20],[10],[-30]])
    prediction=activation(neural_network(np.array([x1,x2,bias]),W_AND))
    if(prnt):
        print(str(x1)+" AND "+ str(x2)+" = "+str(prediction))
    return prediction

#Perceptrón para la compuerta OR
def or_perceptron (x1,x2,prnt=True):
    W_OR=np.array([[20],[20],[-10]])
    prediction=activation(neural_network(np.array([x1,x2,bias]),W_OR))
    if(prnt):
        print(str(x1)+" OR "+ str(x2)+" = "+str(prediction))
    return prediction


#Perceptrón para la compuerta NOT
def not_perceptron (x1,prnt=True):
    W_NOT=np.array([[-20],[0],[10]])
    prediction=activation(neural_network(np.array([x1,x1,bias]),W_NOT))
    if(prnt):
        print("NOT "+ str(x1)+" = "+str(prediction))
    return prediction


#Perceptron para la compuerta XOR
                      
def xor_perceptron(x1,x2):                      
    y1=and_perceptron(x1,x2,False)
    y2=or_perceptron(x1,x2,False)
    y3=not_perceptron(y1,False)
    y4=and_perceptron(y2,y3,False)
    print (str(x1)+" XOR "+str(x2)+" = "+str(y4))
                      
    
print("Calculando valores para la función AND")
and_perceptron(0,0)
and_perceptron(0,1)
and_perceptron(1,0)
and_perceptron(1,1)


print("Calculando valores para la función OR")
or_perceptron(0,0)
or_perceptron(0,1)
or_perceptron(1,0)
or_perceptron(1,1)


print("Calculando valores para la función NOT")
not_perceptron(0)
not_perceptron(1)

print("Calculando valores para la función XOR")    
xor_perceptron(0,0)
xor_perceptron(0,1)
xor_perceptron(1,0)
xor_perceptron(1,1)             
                      
```
